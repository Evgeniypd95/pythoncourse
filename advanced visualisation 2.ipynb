```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
hw = pd.read_csv('/Users/evgenypavlov/Public/Python_projects/Python course/P4-Section6-Homework-Dataset.csv', encoding= 'unicode_escape')
```

```python
hw.columns = ["day_of_week", "director", "genre", "movie_title", "release_date", "studio", "adjusted_gross", "budget", "gross_mill", "imdb", "movielens_rating", "overseas_mill", "overseas_percent", "profit_mill","proft_percent","runtime","us_mill", "gross_percent_us"]
hw.head()
```

```python
genre_list = ["action", "comedy", "adventure", "animation", "drama"]
hw[hw.genre=="action"]
hw2 = hw[(hw.genre=="action") | (hw.genre=="comedy") | (hw.genre=="adventure") | (hw.genre=="animation") | (hw.genre=="drama")]
df = hw2[(hw2.studio== 'Buena Vista Studios') | (hw2.studio== 'Fox')|(hw2.studio== 'Paramount Pictures')|(hw2.studio== 'Sony')|(hw2.studio== 'Universal')|(hw2.studio== 'WB')]
df.head()
# vis1 = sns.boxplot(data=df, x="genre", y="gross_percent_us")
# Define the style
sns.set(style="darkgrid", palette="muted", color_codes=True)

# Plot the boxsplots
ax = sns.boxplot(data=df, x='genre', y='gross_percent_us', orient='v', color='lightgray', showfliers=False)
plt.setp(ax.artists, alpha=0.5)

# Add in points to show each observation
sns.stripplot(x='genre', y='gross_percent_us', data=df, jitter=True, size=6, linewidth=0, hue = 'studio', alpha=0.7)

ax.axes.set_title('chart',fontsize=30)
ax.set_xlabel('genre',fontsize=20)
ax.set_ylabel('gross home',fontsize=20)

# Define where to place the legend
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
